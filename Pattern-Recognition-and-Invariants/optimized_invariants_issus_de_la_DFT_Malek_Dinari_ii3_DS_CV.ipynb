# 1. 1.	Implémenter avec Python les fonctions DFT et DFT-**1**

```python
#  DFT Implementation
def dft(x):
    N = len(x)
    X = np.zeros(N, dtype=complex)
    for k in range(N):
        for n in range(N):
            X[k] += x[n] * np.exp(-2j * np.pi * k * n / N)
    return X
```

```python
#  Inverse DFT Implementation
def idft(X):
    N = len(X)
    x = np.zeros(N, dtype=complex)
    for n in range(N):
        for k in range(N):
            x[n] += X[k] * np.exp(2j * np.pi * k * n / N)
    return x / N
```

```python

#  Example usage
x = np.array([1, 2, 3, 4, 20])  # Example input signal
X = dft(x)
x_reconstructed = idft(X)

print("Input signal:", x)
print("DFT:", X)
print("Reconstructed signal from IDFT:", np.round(x_reconstructed.real, decimals=6))
```

# **a)- Afficher avec Matlab/Python la courbe γ**

# **b)- Calculer F**

```python
import numpy as np

def gamma(t):
    """
    Calculate the parametric curve γ(t) for given time values t.

    Args:
        t (numpy array): Array of time values.

    Returns:
        (x, y) (tuple of numpy arrays): Returns two arrays corresponding to x(t) and y(t).
    """
    #  x(t) and y(t) parametric equations
    x = np.cos(2 * np.pi * t)
    y = 2 * np.sin(2 * np.pi * t) + np.sqrt(np.abs(np.cos(2 * np.pi * t)))

    return x, y

#  Example usage
t = np.arange(0, 1.01, 0.01)  # Time values from 0 to 1 with step 0.01
x, y = gamma(t)

#  Output or plotting (if needed)
print("x(t):", x)
print("y(t):", y)
```

# **c)-	Calculer DFT-1(F) et vérifier que DFT-1(F)=f(t)**

```python
#  Time values from 0 to 1 with step 0.01
t = np.arange(0, 1.01, 0.01)

#  Parametric curve γ(t)
x_t, y_t = gamma(t)

#  Calculate DFT of x(t) and y(t)
X_dft = dft(x_t)
Y_dft = dft(y_t)

#  Calculate inverse DFT (IDFT) of x(t) and y(t)
x_reconstructed = idft(X_dft)
y_reconstructed = idft(Y_dft)

#  Step 51: Plot the original and reconstructed curves

plt.figure(figsize=(10, 8))

#  Original curve
plt.subplot(2, 1, 1)
plt.plot(x_t, y_t, label='Original γ(t)', color='blue')
plt.xlabel('x(t)')
plt.ylabel('y(t)')
plt.title('Original Curve γ(t)')
plt.grid(True)
plt.legend()

#  Reconstructed curve from DFT-1
plt.subplot(2, 1, 2)
plt.plot(x_reconstructed.real, y_reconstructed.real, label='Reconstructed γ(t)', linestyle='--', color='green')
plt.xlabel('x(t)')
plt.ylabel('y(t)')
plt.title('Reconstructed Curve γ(t) from DFT-1')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

#  Print original and reconstructed values for verification
print("Original x(t):", np.round(x_t, decimals=6))
print("Reconstructed x(t):", np.round(x_reconstructed.real, decimals=6))
print("Original y(t):", np.round(y_t, decimals=6))
print("Reconstructed y(t):", np.round(y_reconstructed.real, decimals=6))
```

```python
#  Step 1: Calculate DFT and IDFT for gamma(t)
t = np.arange(0, 1.01, 0.01)
x_t, y_t = gamma(t)

X_dft = dft(x_t)
Y_dft = dft(y_t)

x_reconstructed = idft(X_dft)
y_reconstructed = idft(Y_dft)

#  Step 2: Superimpose Original and Reconstructed Curves
plt.figure(figsize=(8, 6))

#  Superimpose the curves
plt.plot(x_t, y_t, label='Original γ(t)', color='blue')
plt.plot(x_reconstructed.real, y_reconstructed.real, label='Reconstructed γ(t)', linestyle='--', color='red')

plt.xlabel('x(t)')
plt.ylabel('y(t)')
plt.title('Superimposed Original and Reconstructed Curve γ(t)')
plt.grid(True)
plt.legend()
plt.show()

#  Step 3: Calculate Error Metrics
#  Mean Squared Error (MSE)
mse_x = np.mean((x_t - x_reconstructed.real) ** 2)
mse_y = np.mean((y_t - y_reconstructed.real) ** 2)

#  Maximum Absolute Error
max_abs_error_x = np.max(np.abs(x_t - x_reconstructed.real))
max_abs_error_y = np.max(np.abs(y_t - y_reconstructed.real))

#  Print error metrics
print(f"MSE (x): {mse_x:.6f}, MSE (y): {mse_y:.6f}")
print(f"Max Absolute Error (x): {max_abs_error_x:.6f}, Max Absolute Error (y): {max_abs_error_y:.6f}")

#  Step 4: Relative Error
rel_error_x = np.linalg.norm(x_t - x_reconstructed.real) / np.linalg.norm(x_t)
rel_error_y = np.linalg.norm(y_t - y_reconstructed.real) / np.linalg.norm(y_t)

#  Print relative error
print(f"Relative Error (x): {rel_error_x:.6f}, Relative Error (y): {rel_error_y:.6f}")
```

# **3. 	Soit f_2 (t)=a f(t).  Afficher dans la même figure f_2 et f et dans une autre la variation de I_n (f) et I_n (f_2)**

```python
import numpy as np
import matplotlib.pyplot as plt

#  Step 1: Définir the Gamma fonction (as a complex fonction)
    x = np.cos(2 * np.pi * t)
    y = 2 * np.sin(2 * np.pi * t) + np.sqrt(np.abs(np.cos(2 * np.pi * t)))
    return x + 1j * y  # Return as a complex function

#  Step 2: Définir the DFT fonction
    N = len(x)
    X = np.zeros(N, dtype=complex)
    for k in range(N):
        for n in range(N):
            X[k] += x[n] * np.exp(-2j * np.pi * k * n / N)
    return X

#  Step 3: Définir the Fourier invariant calculation
def calculate_fourier_invariants(X):
    X_norm = np.abs(X)
    I_n = X_norm / X_norm[2]  # Division by the norm of the 2nd DFT component
    return I_n

#  Step 4: Generate the time samples
t = np.arange(0, 1.01, 0.01)
f = gamma(t)  # f(t) is now a complex function

#  Step 5: Définir the scaling factor a
a = 0.5  # You can change this value as needed
f_2 = a * f  # f_2 is also a complex function

#  Step 6: Calculer DFTs for both f and f_2
F = dft(f)
F_2 = dft(f_2)

#  Step 7: Calculate the invariants for f and f_2
I_f = calculate_fourier_invariants(F)
I_f_2 = calculate_fourier_invariants(F_2)

#  Step 8: Plot f and f_2 in the same figure
plt.figure(figsize=(10, 6))
plt.plot(f.real, f.imag, label='f(t) = γ(t)', color='blue')
plt.plot(f_2.real, f_2.imag, label='f_2(t) = a * γ(t)', linestyle='--', color='green')
plt.xlabel('Re(f(t))')
plt.ylabel('Im(f(t))')
plt.title('Comparaison de f(t) et f_2(t)')
plt.grid(True)
plt.legend()
plt.axis('equal')  # To maintain aspect ratio
plt.show()

#  Step 9: Plot the variation of I_n(f) and I_n(f_2)
n_values = np.arange(len(I_f))  # Corresponds to the DFT index
plt.figure(figsize=(10, 6))
plt.plot(n_values, I_f, label='I_n(f)', color='blue')
plt.plot(n_values, I_f_2, label='I_n(f_2)', linestyle='--', color='red')
plt.xlabel('n')
plt.ylabel('Invariant de Fourier I_n')
plt.title('Variation de I_n(f) et I_n(f_2)')
plt.grid(True)
plt.legend()
plt.show()
```
